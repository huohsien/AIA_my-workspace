```python
def fibonacci(max_number):
    pre_v = 0
    pre_pre_v = 0
    v = 0
    
    if max_number < 0:
        return
    
    if max_number == 0:
        yield 0
        return
    elif max_number == 1:
        yield 0
        yield 1
        yield 1
        return
    yield 0
    yield 1
    pre_pre_v = 0
    pre_v = 1
    v = 1
    while(v < max_number):
        yield v
        pre_pre_v = pre_v
        pre_v = v
        v = pre_pre_v + pre_v

    return

# testing code

fib_gen = fibonacci(50)

for i in range(10000):
    print(next(fib_gen))
```

```python
# Q1
class Rectangle:
    width = 0.0
    height = 0.0

    def area(self):
        return self.height * self.width
    
# testing code
#

my_rect = Rectangle()
my_rect.width = 10
my_rect.height = 20.5

print("area = ", my_rect.area())

# Q2

class Student:
    def calculate(self, rect1, rect2):
        return rect1.area() + rect2.area()

# Q3
#

stu = Student()
rect1 = Rectangle()
rect1.width = 2
rect1.height = 3
rect2 = Rectangle()
rect2.width = 3
rect2.height = 4

print("sum of the area of the two rectangles:", stu.calculate(rect1, rect2))
        
```
